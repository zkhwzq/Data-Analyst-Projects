# Data Import

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads=pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
recent_grsds=recent_grads.dropna()
```

# Relation between Sample_size and Unemployment_rate

# Histograms Investigation

```python

cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
fig=plt.figure(figsize=(5,5*len(cols)))
for index, item in enumerate(cols):
    ax=fig.add_subplot(len(cols),1,index+1)
    ax=recent_grads[item].plot(kind='hist')
    ax.set_title('Histogram of ' + item)
```
